Make a convenience function for calculating the magnetic field in three slices in the xy, xz, yz planes

```python
def plot_sphere_example(alpha=0, beta=0, gamma=0, num_levels =6, cmax = 0.5, mask_magnet=False):
    pm.reset_magnets();
    cmap = 'viridis'

    NP = 201
    radius = 10e-3
    Jr = 1.0
    xc = radius
    yc = radius/2
    zc = -radius/2
    center = (0,0,0)
    NN = 4


    s1 = pm.magnets.Sphere(radius=radius, Jr=1.0, center=center,
                           alpha=alpha,
                           beta=beta,
                           gamma=gamma,
                           mask_magnet=mask_magnet)
    

    x, z = pm.grid2D(NN*radius, NN*radius, NP=NP)
    y = np.array([0])
    B = pm.B_calc_3D(x, y, z)
    B.calc_norm()
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)


    x, y = pm.grid2D(NN*radius, NN*radius, NP=NP)
    z = np.array([0])
    B = pm.B_calc_3D(x, y, z)
    B.calc_norm()

    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)

    y, z = pm.grid2D(NN*radius, NN*radius, NP=NP)
    x = np.array([0])
    B = pm.B_calc_3D(x, y, z)
    B.calc_norm()

    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)


    return s1
```

```python
mag1 = plot_sphere_example(alpha=0, beta=0, gamma=0, num_levels =6, cmax = 0.5, mask_magnet=True)
```

```python
def rotate_prism(alpha=0.0, beta=0.0, gamma=0.0, mask_magnet=False):
    pm.reset_magnets()
    width = 20e-3
    depth = 40e-3
    height = 60e-3
    center = (0, 0, 0)

    # magnetised in x: 
    # theta, phi = 0.0, 90.0

    # magnetised in y: 
    # theta, phi = 90.0, 90.0


    # magnetised in z: 
    theta, phi = 45.0, 45.0
    # or
    # theta, phi= 0.0, 0.0
    cmap = 'viridis'
    num_levels = 13
    cmax=0.6
    magnet = pm.magnets.Prism(width = width, depth = depth, height = height, Jr = 1.0, center=center,
                              theta = theta, phi = phi,
                              alpha = alpha,
                              beta = beta,
                              gamma = gamma,
                              mask_magnet=mask_magnet)


    x, z = pm.grid2D(height, height)
    y = 0

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)



    x, y = pm.grid2D(height, height)
    z = height*1.001/3

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)





    y, z = pm.grid2D(height, height)
    x = 0 

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)

```

```python
def rotate_cube(alpha=0.0, beta=0.0, gamma=0.0, mask_magnet=False):
    pm.reset_magnets()
    a = 20e-3
    b = 40e-3
    c = 60e-3
    center = (0, 0, 0)

    # magnetised in x: 
    # theta, phi = 0.0, 90.0

    # magnetised in y: 
    # theta, phi = 90.0, 90.0


    # magnetised in z: 
    theta, phi = 90.0, 0.0
    # or
    # theta, phi= 0.0, 0.0
    cmap = 'viridis'

    magnet = pm.magnets.Cube(width = a, Jr = 1.0, center=center,
                              theta = theta, phi = phi,
                              alpha = alpha,
                              beta = beta,
                              gamma = gamma,
                              mask_magnet=mask_magnet)


    x, z = pm.grid2D(c, c)
    y = 0

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)



    x, y = pm.grid2D(c, c)
    z = a*1.001/2

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)





    y, z = pm.grid2D(c, c)
    x = 0 

    B = pm.B_calc_3D(x, y, z)
    pm.plots.plot_3D_contour(x, y, z, B,
                             cmax=cmax, num_levels=num_levels, cmap = cmap,
                             vector_plot=True,
                             vector_arrows=11)


```

```python
def rotate_cylinder(alpha=0.0, beta=0.0, gamma=0.0,mask_magnet=False):
    pm.reset_magnets()
    
    center = (0, 0, 0)
    R = 5e-3
    L = 20e-3
    
    num_levels = 13
    cmax=0.6

    m_cyl = pm.magnets.Cylinder(radius = R, length = L, Jr = 1.0,
                                center=center,
                                theta = 0, phi = 0,
                                alpha = alpha,
                                beta = beta,
                                gamma = gamma,
                                mask_magnet=mask_magnet)
    



    x, z = pm.grid2D(R*3, L*1.5)
    y = 0
    
    B = pm.B_calc_3D(x, y, z)

    cmap = 'viridis'
    _ = pm.plots.plot_3D_contour(x, y, z, B, cmax=1, num_levels=11, cmap = cmap)
    B2 = pm.magnets._fields.Vector2(B.x, B.z)
    B2.calc_norm()
    pm.plots._plot2._vector_plot2(x, z,  B2, NQ = 11, scale_x=1e-3, scale_y=1e-3, vector_color='w')




    x, y = pm.grid2D(R*3, R*3)
    z = L/2
    
    B = pm.B_calc_3D(x, y, z)
    _ = pm.plots.plot_3D_contour(x, y, z, B, cmax=1, num_levels=11, cmap = cmap)
    B2 = pm.magnets._fields.Vector2(B.x, B.y)
    B2.calc_norm()
    pm.plots._plot2._vector_plot2(x, y,  B2, NQ = 11, scale_x=1e-3, scale_y=1e-3, vector_color='w')





    y, z = pm.grid2D(R*3, L*1.5)
    x = 0 

    B = pm.B_calc_3D(x, y, z)
    _ = pm.plots.plot_3D_contour(x, y, z, B, cmax=1, num_levels=11, cmap = cmap)
    B2 = pm.magnets._fields.Vector2(B.y, B.z)
    B2.calc_norm()
    pm.plots._plot2._vector_plot2(y, z,  B2, NQ = 11, scale_x=1e-3, scale_y=1e-3, vector_color='w')

    return m_cyl
```
