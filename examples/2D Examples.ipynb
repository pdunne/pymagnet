```python
# Matplotlib settings
%matplotlib inline
# %matplotlib notebook
%config InlineBackend.figure_format = 'retina'
import pymagnet as pm
```

## Rectangles

```python
pm.reset_magnets() # clear magnet registry

cmap = 'viridis' # set the colormap (optional)

width = 20
height = 20
unit = 'mm'
# Create the magnet
m_rectangle = pm.magnets.Rectangle(width=width,
                         height=height,
                         Jr=1.0, # Remnant magnetisation in T
                        )
```

```python
m_rectangle
```

```python
# Prepare 100x100 grid of x,y coordinates to calculate the field
points = pm.grid2D(2 * width, 2 * height)

# Calculate the magnetic field due to all magnets in the registry
field = pm.B_calc_2D(points)

# Plot the result, vector_plot = True toggles on the vector field plot
pm.plots.plot_2D_contour(points, field, cmin=0.0, # minimum field value
                                 cmax=0.5, # maximum field value
                                 vector_plot=True, # plot the vector field
                                 cmap=cmap, # set the colormap
#                                  show_magnets=False,
                                 )

pm.plots.plot_2D_contour(points, field, cmin=-0.1, # minimum field value
                                 cmax=0.1, # maximum field value
                                 cmap='coolwarm', # set the colormap
                                 plot_type='streamplot',
                                 show_magnets=True,
                        )
```

## Circular Magnets

```python
pm.reset_magnets();
cmap = 'viridis'

radius = 10

center = (0, 0)
_ = pm.magnets.Circle(radius=radius, Jr = 1.0, center=center, alpha=45)


points = pm.grid2D(2*radius, 2*radius)
field = pm.B_calc_2D(points)

# pm.plots.plot_2D_contour(points, field, cmax=0.5, num_lines=6, cmap=cmap, vector_plot=True, vector_arrows=11)

# Plot the result, vector_plot = True toggles on the vector field plot
pm.plots.plot_2D_contour(points, field,
                         cmax=0.5,
                         num_levels=6,
                         cmap=cmap,
                         vector_plot=True,
                         vector_arrows=11)



# Plot the result as a streamplot 
pm.plots.plot_2D_contour(points, field,
                         cmin = -0.3,
                         cmax=0.3,
                         cmap='coolwarm',
                         plot_type="streamplot",
                         stream_color= 'vertical', # 'vertical', 'horizontal', 'normal':
#                          corresponds to coloring by B.x, B.y, B.n
                        )

# If cmap is not defined, a black streamplot is drawn:
pm.plots.plot_2D_contour(points, field,
                         plot_type="streamplot",
                         color = 'black', # default is black
                        )
```
